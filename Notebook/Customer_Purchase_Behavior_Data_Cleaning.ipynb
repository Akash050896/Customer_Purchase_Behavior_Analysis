```python
# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the Item Purchased

item_median = df.groupby('Item Purchased')['Review Rating'].transform('median')

df['Review Rating'] = df['Review Rating'].fillna(item_median)
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
bins = [18, 30, 45, 60, 100]
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
#  Age     Age	Groups
# 18–29	   Young Adult
# 30–44	   Adult
# 45–59	   Middle-Aged
# 60+	   Senior

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)


```

```python
df[['age','age_group']].head(10)
```

```python
print(df['frequency_of_purchases'].unique())
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
    'Annually': 365
    }

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
